```python
'''
Credit to Brendan Artley for code on colour gradient in matplotlib
https://medium.com/@BrendanArtley/matplotlib-color-gradients-21374910584b

'''

def hex_to_RGB(hex_str):
    """ #FFFFFF -> [255,255,255]"""
    #Pass 16 to the integer function for change of base
    return [int(hex_str[i:i+2], 16) for i in range(1,6,2)]

def get_color_gradient(c1, c2, n):
    """
    Given two hex colors, returns a color gradient
    with n colors.
    """
    assert n > 1
    c1_rgb = np.array(hex_to_RGB(c1))/255
    c2_rgb = np.array(hex_to_RGB(c2))/255
    mix_pcts = [x/(n-1) for x in range(n)]
    rgb_colors = [((1-mix)*c1_rgb + (mix*c2_rgb)) for mix in mix_pcts]
    return ["#" + "".join([format(int(round(val*255)), "02x") for val in item]) for item in rgb_colors]

#color=get_color_gradient(c1,c2,samples)
```

```python

l = list(b['original_publication_year'])
f = open('t.txt','w')
for i in l:
    f.write(str(i)+',')
f.close()
f = open('t.txt','r')
l = f.read().split(',')
no_of_books_published_in_a_year = dict()
nan_index = []
for i in range(len(l)):
    if l[i] == 'nan':
        nan_index.append(i)
    if l[i] not in no_of_books_published_in_a_year.keys():
        no_of_books_published_in_a_year[l[i]] = 1
    else:
        no_of_books_published_in_a_year[l[i]] += 1

no_of_nan = no_of_books_published_in_a_year.pop('nan')
bakwas = no_of_books_published_in_a_year.pop('')
range_of_years = [i for i in range(-1750,2150,100)]
no_of_books_published_in_a_century = dict()
for i in range_of_years:
    no_of_books_published_in_a_century[i] = 0

for i in no_of_books_published_in_a_year.keys():
    for j in range_of_years:
        if int(float(i))>=j and int(float(i))<j+100:
            if abs(int(float((i)))-j)<abs(int(float((i)))-(j+100)):
                no_of_books_published_in_a_century[j] += no_of_books_published_in_a_year[i]
            else:
                no_of_books_published_in_a_century[j+100] += no_of_books_published_in_a_year[i]
            break
years = []
for i in range_of_years:
    if i == -50:
        years.append('0')
    elif i<0:
        years.append(f'{(-i-50)} BC')
    else:
        years.append(f'{i+50} AD')

bc = b.copy()
bc = bc.sort_values(by='original_publication_year')
list_of_indices = []
for i in range(len(bc)):
    if i in nan_index:
        continue
    if int(bc['original_publication_year'][i]) < 1800:
        list_of_indices.append(i)
bc = bc.drop(list_of_indices+nan_index,axis=0)
no_of_books_published_in_a_year_after_1800 = dict()
for i in bc['original_publication_year']:
    if i not in no_of_books_published_in_a_year_after_1800.keys():
        no_of_books_published_in_a_year_after_1800[int(i)] = 0
    else:
        no_of_books_published_in_a_year_after_1800[int(i)] += 1

fig,ax = plt.subplots(1,2,figsize=(16,5))
plt.subplot(1,2,1)
plt.bar(no_of_books_published_in_a_century.keys(),no_of_books_published_in_a_century.values(),width=90,color='r')
plt.bar(1760,7000,width=20)
plt.xticks(list(no_of_books_published_in_a_century.keys()),labels=years,rotation=90)
plt.ylim((-100,6300))
plt.xlabel('Century')
plt.ylabel('Number of books published')
plt.title('Total number of books published in a century')
plt.grid(axis = 'y')
plt.legend(['Number of books published','90th prcentile'])
plt.subplot(1,2,2)
plt.bar(no_of_books_published_in_a_year_after_1800.keys(),no_of_books_published_in_a_year_after_1800.values(),color=get_color_gradient('#0000ff','#00ff00',len(no_of_books_published_in_a_year_after_1800)))
plt.title('Total number of books published in a year after 1800 AD') # Max 567 in 2012
plt.ylabel('Number of books published')
plt.xlabel('Years')
plt.grid(axis='y')
plt.show()
```

```python
b_sorted_by_best_book_id = bc.copy()
b_sorted_by_best_book_id = b_sorted_by_best_book_id.sort_values(by='best_book_id')

fig,ax = plt.subplots(2,2,figsize=(20,10))
plt.subplot(2,2,1)
plt.scatter(b_sorted_by_best_book_id['best_book_id'],b_sorted_by_best_book_id['weighted_average_of_ratings'],color=get_color_gradient('#c0c0c0','#000000',len(b_sorted_by_best_book_id)))
plt.title('Scatter plot of average rating of best book IDs')
plt.xlabel('Best book IDs')
plt.ylabel('Average rating')
plt.grid(axis='y',color=(0,0,0))

plt.subplot(2,2,2)
b_sorted_by_best_book_id['weighted_average_of_ratings'].plot.kde(bw_method=0.1)
plt.title('Density of average rating of best book IDs')
plt.xlabel('Average rating')
plt.tick_params(labelleft=False,left=False)

plt.subplot(2,2,3)
plt.scatter(b_sorted_by_best_book_id['best_book_id'],b_sorted_by_best_book_id['work_ratings_count'],color=get_color_gradient('#c0c0c0','#000000',len(b_sorted_by_best_book_id)))
plt.title('Scatter plot of work ratings count of best book IDs')
plt.xlabel('Best book IDs')
plt.ylabel('Work ratings count')
plt.grid(axis='y')

plt.subplot(2,2,4)
b_sorted_by_best_book_id['work_ratings_count'].plot.kde(bw_method=0.1)
plt.title('Density of work ratings count of best book IDs')
plt.xlabel('Work ratings count')
plt.xlim((0,1.5*1000000))
plt.tick_params(labelleft=False,left=False)

plt.show()
```
