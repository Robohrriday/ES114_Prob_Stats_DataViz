```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('AusOpen-men-2013.csv')
display(df)
```

```python
tournaments_removals = {'AusOpen-men-2013':['Player1','Player2'],
 'AusOpen-women-2013':['Player1','Player2','ST4.1','ST5.1','ST4.2','ST5.2'],
 'FrenchOpen-men-2013.csv':['Player1','Player2'],
 'FrenchOpen-women-2013.csv':['Player1','Player2','ST4.1','ST5.1','ST4.2','ST5.2'],
 'USOpen-men-2013':['Player1','Player2','WNR.1','UFE.1','WNR.2','UFE.2'],
 'USOpen-women-2013':['Player 1','Player 2','ST4.1','ST5.1','ST4.2','ST5.2','TPW.1','TPW.2'],
 'Wimbledon-men-2013.csv':['Player1','Player2','TPW.1','TPW.2'],
 'Wimbledon-women-2013.csv':['Player1','Player2','ST4.1','ST5.1','ST4.2','ST5.2','TPW.1','TPW.2'],
}

df1 = df.drop(tournaments_removals['AusOpen-men-2013'], axis=1)
corr_matrix = df1.corr()
plt.figure(figsize=(25,25))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True)
plt.show()


def give_weights(tournament,tournaments_removals,df):
    df1 = df.drop(tournaments_removals[tournament], axis=1)
    corr_matrix = df1.corr()
    corr_threshold = 0.5
    important_features_pcorr = corr_matrix['Result'][corr_matrix['Result'] > corr_threshold].index.tolist()
    important_features_ncorr = corr_matrix['Result'][corr_matrix['Result'] < -corr_threshold].index.tolist()
    print("Important features Positive Correlation:", important_features_pcorr)
    print("Important features Negative Correlation:", important_features_ncorr)
    return list(corr_matrix['Result'])

weights = give_weights('AusOpen-men-2013',tournaments_removals,df)
print(weights)
```

```python
def calculate_performance_index(df, player_name,tournament,tournaments_removals):
    player_df = df[(df['Player1'] == player_name) | (df['Player2'] == player_name)]
    total_matches = len(player_df)
    wins = len(player_df[((player_df['Player1'] == player_name) & (player_df['Result']==1)) | ((player_df['Player2'] == player_name) & (player_df['Result']==0)) ])
    win_ratio = wins / total_matches

    features = ['FSP.1', 'FSW.1', 'SSP.1', 'SSW.1', 'ACE.1', 'DBF.1', 'WNR.1', 'UFE.1', 'BPC.1', 'BPW.1', 'NPA.1', 'NPW.1', 'TPW.1','ST1.1','ST2.1','ST3.1','ST4.1','ST5.1',
                'FSP.2', 'FSW.2', 'SSP.2', 'SSW.2', 'ACE.2', 'DBF.2', 'WNR.2', 'UFE.2', 'BPC.2', 'BPW.2', 'NPA.2', 'NPW.2', 'TPW.2','ST1.2','ST2.2','ST3.2','ST4.2','ST5.2']
    player_stats = player_df[features].mean()
    s = 0
    weights = give_weights(tournament,tournaments_removals,df)
    for i in range(len(features)):
        s += player_stats[i]*weights[4+i]
    performance_index = s/np.sum(np.array(weights))
    return performance_index,win_ratio

df2 = df.fillna(0)
player_performance = {}
players = list(set(list(df2['Player1'].unique()) + list(df2['Player2'].unique())))
for player_name in players:
    performance_index = calculate_performance_index(df2,player_name,'AusOpen-men-2013',tournaments_removals)
    player_performance[player_name] = performance_index[0]

sorted_performance = dict(sorted(player_performance.items(), key=lambda item: item[1], reverse=True))

print("Player rankings by performance index:")
for i, (player_name, performance_index) in enumerate(sorted_performance.items()):
    print(f"{i+1}. {player_name}: {performance_index:.2f}")
```
