```python
df = df.fillna(0)
```

```python
x = X
y = Y

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
hist, xedges, yedges = np.histogram2d(x, y, bins=50)
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0
dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()
ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='r', alpha=0.5,edgecolor = 'white')
ax.view_init(elev=60, azim=-60)
ax.set_xlabel('X',fontsize = 20)
ax.set_ylabel('Y',fontsize = 20)
ax.set_zlabel('Frequency',fontsize = 20)
ax.set_title('Joint Distribution of X and Y',fontsize = 30)
plt.show()
```
