```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
pd.options.display.max_columns = None
```

```python
# Aus Open = Hard Court
# French Open = Clay Court
# Wim = Grass Court
# US Open = Hard Court
df = pd.read_csv('Wimbledon-men-2013.csv')
display(df)
```

```python
df1 = df[(df['Player1'] == 'N.Djokovic') | (df['Player2'] == 'N.Djokovic')]


df2 = df[(df['Player1'] == 'A.Murray') | (df['Player2'] == 'A.Murray')]


display(df2)
plt.plot(df1['Round'],df1['BPC.1'],label='ND BPC.1')
plt.plot(df1['Round'],df1['BPW.1'],label='ND BPW.1')
plt.plot(df2['Round'],df2['BPC.2'],label='AM BPC.2')
plt.plot(df2['Round'],df2['BPW.2'],label='AM BPW.2')
plt.title('Performance Progress')
plt.xlabel('Rounds')
plt.ylabel('Count')
plt.legend()
plt.show()
```
