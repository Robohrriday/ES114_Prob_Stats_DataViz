```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
pd.options.display.max_columns = None
```

```python
# Aus Open = Hard Court
# French Open = Clay Court
# Wim = Grass Court
# US Open = Hard Court
df = pd.read_csv('Wimbledon-women-2013.csv')
display(df)
```

```python
df1 = df[(df['Player1'] == 'S.Lisicki') | (df['Player2'] == 'S.Lisicki')]

df2 = df[(df['Player1'] == 'M.Bartoli') | (df['Player2'] == 'M.Bartoli')]
df1 = df1.reset_index()
df2 = df2.reset_index()

display(df2)
plt.plot(df1['Round'], list(df1['DBF.1'])[0:3] + [list(df1['DBF.2'])[3]] + list(df1['DBF.1'])[4:7] ,label='SL DBF.1')
plt.plot(df1['Round'],list(df1['UFE.1'])[0:3] + [list(df1['UFE.2'])[3]] + list(df1['UFE.1'])[4:7] ,label='SL UFE.1')
plt.plot(df2['Round'],list(df2['DBF.1'])[0:4] + [list(df2['DBF.2'])[4]] + [list(df2['DBF.1'])[5]] + [list(df2['DBF.2'])[6]],label='MB DBF.2')
plt.plot(df2['Round'],list(df2['UFE.1'])[0:4] + [list(df2['UFE.2'])[4]] + [list(df2['UFE.1'])[5]] + [list(df2['UFE.2'])[6]],label='MB UFE.2')
plt.xlabel('Rounds')
plt.ylabel('Count')
plt.title('Error Progress')
plt.legend()
plt.show()
```
